```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('country_wise_latest.csv')   # change filename if different

# Get top 10 countries by Confirmed cases
top10 = df.sort_values('Confirmed', ascending=False).head(10)

# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(top10['Country/Region'], top10['Confirmed'], color='skyblue')
plt.title('Top 10 Countries by Confirmed COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('country_wise_latest.csv')   # change filename if different

# Total Confirmed cases grouped by WHO Region
region_data = df.groupby('WHO Region')['Confirmed'].sum()

# Pie Chart
plt.figure(figsize=(7, 7))
plt.pie(region_data, labels=region_data.index, autopct='%1.1f%%', startangle=90)
plt.title('Confirmed COVID-19 Cases by WHO Region')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('country_wise_latest.csv')   # change filename if different

# Histogram of Death Rate (Deaths per 100 Cases) across countries
plt.figure(figsize=(8, 6))
plt.hist(df['Deaths / 100 Cases'], bins=10, color='salmon', edgecolor='black')
plt.title('Distribution of Death Rate Across Countries')
plt.xlabel('Deaths / 100 Cases')
plt.ylabel('Number of Countries')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('country_wise_latest.csv')   # change filename if different

# Scatter Plot: Confirmed vs Deaths
plt.figure(figsize=(8, 6))
plt.scatter(df['Confirmed'], df['Deaths'], color='purple')
plt.title('Confirmed Cases vs Deaths (Country-wise)')
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('country_wise_latest.csv')   # change filename if different

# Take a small sample - just 10 countries' Confirmed cases
sample_data = df['Confirmed'].head(10)
print(sample_data)

# Simple Box Plot with labels
plt.boxplot(sample_data, labels=['Confirmed Cases'])
plt.title('Box Plot of Confirmed Cases (Sample)')
plt.xlabel('Category')
plt.ylabel('Confirmed Cases')

# Print the key values so you can label them manually / understand the plot
print("Lowest sample:", sample_data.min())
print("Lower Quartile (Q1):", sample_data.quantile(0.25))
print("Median:", sample_data.median())
print("Upper Quartile (Q3):", sample_data.quantile(0.75))
print("Highest sample:", sample_data.max())

plt.show()
```
