````python
### Differences between List, Tuple, and Dictionary in Python

#### **1. List**
-   **Definition**: An ordered, mutable (changeable) collection of items.
-   **Syntax**: Defined using square brackets `[]`.
-   **Mutability**: Elements can be added, removed, or modified after creation.
-   **Ordered**: Maintains the order of insertion.
-   **Indexed**: Elements are accessed by integer indices.
-   **Heterogeneous**: Can contain items of different data types.
-   **Use Cases**: When you need a collection of items that can change, grow, or shrink dynamically.
-   **Example**:
    ```python
    my_list = [1, "hello", 3.14, True]
    my_list.append(5)
    my_list[0] = 10
    ```

#### **2. Tuple**
-   **Definition**: An ordered, immutable (unchangeable) collection of items.
-   **Syntax**: Defined using parentheses `()`.
-   **Mutability**: Elements cannot be added, removed, or modified after creation. Once a tuple is created, its contents are fixed.
-   **Ordered**: Maintains the order of insertion.
-   **Indexed**: Elements are accessed by integer indices.
-   **Heterogeneous**: Can contain items of different data types.
-   **Use Cases**: When you need a fixed collection of items, or as dictionary keys (since they are immutable) or function arguments that should not be altered.
-   **Example**:
    ```python
    my_tuple = (1, "hello", 3.14, True)
    # my_tuple.append(5)  # This would cause an error
    # my_tuple[0] = 10    # This would cause an error
    ```

#### **3. Dictionary**
-   **Definition**: An unordered (in Python versions before 3.7) or ordered (Python 3.7+), mutable collection of key-value pairs.
-   **Syntax**: Defined using curly braces `{}` with key-value pairs separated by colons `:`.
-   **Mutability**: Key-value pairs can be added, removed, or modified after creation. Values associated with keys can be changed.
-   **Ordered**: In Python 3.7+, dictionaries maintain insertion order.
-   **Key-Value Pairs**: Each item consists of a key and its associated value. Keys must be unique and immutable (e.g., strings, numbers, tuples).
-   **Access**: Elements are accessed by their keys, not by integer indices.
-   **Use Cases**: When you need to store data in a way that you can quickly look up values based on a unique identifier (key).
-   **Example**:
    ```python
    my_dict = {"name": "Alice", "age": 30, "is_student": False}
    my_dict["city"] = "New York"
    my_dict["age"] = 31
    ```

#### **Summary Table**

| Feature        | List                 | Tuple                  | Dictionary          |
|----------------|----------------------|------------------------|---------------------|
| **Mutability** | Mutable              | Immutable              | Mutable             |
| **Order**      | Ordered              | Ordered                | Ordered (Python 3.7+)|
| **Syntax**     | `[]`                 | `()`                   | `{}`                |
| **Access**     | Index (`my_list[0]`) | Index (`my_tuple[0]`)  | Key (`my_dict["key"]`)|
| **Purpose**    | Collection of items  | Fixed collection of items| Key-value mapping  |
````
